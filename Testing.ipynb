```python
import pandas as pd
import geopandas as gpd

# Load the Excel file
df_lstat = pd.read_excel("datasets/Ladesaeulenregister_SEP.xlsx", header=10)

# Extract relevant columns (adjust column names as needed)
# df_lstat = df_lstat[['PLZ', 'Breitengrad', 'Längengrad']]  # Replace with actual column names

# Count the number of charging stations per PLZ
lstat_counts = df_lstat.groupby('PLZ').size().reset_index(name='count')
df_lstat_2 = df_lstat[["PLZ", "Ort"]]

# Convert PLZ columns to strings to ensure proper merge
lstat_counts['PLZ'] = lstat_counts['PLZ'].astype(str)
df_lstat_2['PLZ'] = df_lstat_2['PLZ'].astype(str)

# Merge the lstat counts with df_lstat_2
merged_df = pd.merge(lstat_counts, df_lstat_2, on='PLZ', how='inner')

# Load the GeoJSON file
gdf = gpd.read_file("output.geojson")

# Convert PLZ columns to strings to ensure proper merge
gdf['PLZ'] = gdf['PLZ'].astype(str)

# Merge the GeoJSON data with the merged DataFrame
gdf = gdf.merge(merged_df, on='PLZ', how='left')

# Check the result of the merge
gdf.to_file("output.geojson", driver="GeoJSON")

print(gdf.head())
```

```python
import pandas as pd
import geopandas as gpd

# Load the Excel file
df_lstat = pd.read_excel("datasets/Ladesaeulenregister_SEP.xlsx", header=10)

# Extract relevant columns (adjust column names as needed)
df_lstat = df_lstat[['PLZ', 'Breitengrad', 'Längengrad']]  # Replace with actual column names

# Count the number of charging stations per PLZ
lstat_counts = df_lstat.groupby('PLZ').size().reset_index(name='count')

# Convert PLZ columns to strings to ensure proper merge
lstat_counts['PLZ'] = lstat_counts['PLZ'].astype(str)

# Load the GeoJSON file
gdf = gpd.read_file("output.geojson")

# Convert PLZ columns to strings to ensure proper merge
gdf['PLZ'] = gdf['PLZ'].astype(str)

# Print the first few rows of each DataFrame to inspect the data
print(lstat_counts.head())
print(gdf.head())

# Merge the GeoJSON data with the charging station counts
gdf = gdf.merge(lstat_counts, on='PLZ', how='left')

# Check the result of the merge
gdf.to_file("output.geojson", driver="GeoJSON")
```
